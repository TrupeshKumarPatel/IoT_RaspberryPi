```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
log_file_path = "../../experiment_results/"
log_list = os.listdir(log_file_path)
print(log_list)
```

```python
## Extract results from "diff_place_sec_to_bus_hub_2.log" 
each_log_path = os.path.join(log_file_path, log_list[0])
df_bus_hub_2 = pd.DataFrame(columns = ['Peer', 'Recv:Pack', 'Recv:Lost%', 'Recv:Jitter', 
                                       'Send:Pack', 'Send:Lost%', 'Send:Jitter'])

f = open(each_log_path, "r")
for each_line in f:
    line_split = each_line.split()
    if (len(line_split) == 13):
        df_each_line = pd.DataFrame([[line_split[0], line_split[3], line_split[6].replace('%)',''), line_split[7],
                                      line_split[8], line_split[11].replace('%)',''), line_split[12]]], 
                                    columns = ['Peer', 'Recv:Pack', 'Recv:Lost%', 'Recv:Jitter', 
                                               'Send:Pack', 'Send:Lost%', 'Send:Jitter'])
        df_bus_hub_2 = pd.concat([df_bus_hub_2, df_each_line])
        
print("="*70)
print("Results at : ", log_list[0])
print("="*70)
print(df_bus_hub_2)
```
